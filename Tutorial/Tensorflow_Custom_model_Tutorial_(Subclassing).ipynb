```python
from tensorflow.keras import layers
import tensorflow as tf
```

```python
class Custom_model(tf.keras.Model):

  def __init__(self, hidden_dimension1, hidden_dimension2, hidden_dimension3, hidden_dimension4, \
               hidden_dimension5, hidden_dimension6, output_dimension):
    super(Custom_model, self).__init__(name='model')

    self.dense_layer1 = layers.Dense(hidden_dimension1, input_shape=(7,), activation='relu')
    self.dense_layer2 = layers.Dense(hidden_dimension2, activation='relu')
    self.dense_layer3 = layers.Dense(hidden_dimension3, activation='relu')
    self.dense_layer4 = layers.Dense(hidden_dimension4, activation='relu')
    self.dense_layer5 = layers.Dense(hidden_dimension5, activation='relu')
    self.dense_layer6 = layers.Dense(hidden_dimension6, activation='relu')
    self.dense_layer7 = layers.Dense(output_dimension, activation='softmax')

  def call(self, inputs):
    x = self.dense_layer1(inputs)
    x = self.dense_layer2(x)
    x = self.dense_layer3(x)
    x = self.dense_layer4(x)
    x = self.dense_layer5(x)
    x = self.dense_layer6(x)

    return self.dense_layer7(x)
```

```python
from tensorflow import keras
from tensorflow.keras.utils import to_categorical
from imblearn.over_sampling import SMOTE

import pandas as pd

dataset_path = keras.utils.get_file("auto-mpg.data", "http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data")

column_names = ['MPG','Cylinders','Displacement','Horsepower','Weight',
                'Acceleration', 'Model Year', 'Origin']
raw_dataset = pd.read_csv(dataset_path, names=column_names,
                      na_values = "?", comment='\t',
                      sep=" ", skipinitialspace=True)

dataset = raw_dataset.copy()
dataset = dataset.dropna()

x = dataset.iloc[:,:7]
y = dataset.iloc[:,7:]

sm = SMOTE(sampling_strategy='auto')

x_train, y_train = sm.fit_resample(x, y)

print(x_train.shape)
print(y_train.shape)

resampled_dataset = pd.concat([x_train, y_train], axis=1)
resampled_dataset.tail()
```

```python
origin = resampled_dataset.pop('Origin')

resampled_dataset['origin_1'] = (origin == 1)*1.0
resampled_dataset['origin_2'] = (origin == 2)*1.0
resampled_dataset['origin_3'] = (origin == 3)*1.0

resampled_dataset.tail()
```

```python
x = resampled_dataset.iloc[:,:7]
y = resampled_dataset.iloc[:,7:]

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=121)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = Custom_model(1024, 512, 256, 3)
```

```python
model.compile(optimizer='Adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

model.fit(x_train, y_train,
          batch_size=20,
          epochs=500,
          callbacks=[EarlyStopping(monitor='loss', patience=10)])
```
